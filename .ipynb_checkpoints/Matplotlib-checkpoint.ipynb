```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = 2*x
plt.title("First Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0,10)
plt.ylim(0,20)
plt.plot(x,y)
# plt.savefig("firstgraph.png")
```

## Figure Object

```python
# Figures and Axes
a = np.linspace(0,10,11)
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
y = x * 2
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')
axes1.plot(a,b)

# SMALL AXES
axes2 = fig.add_axes([.2,.2, 0.25,0.25])
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
axes2.plot(a,b)


plt.show()
```

## Figure Parameters

```python
fig = plt.figure(figsize=(2,2), dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
#fig.savefig('EX_figure_params', bbox_inches='tight')
plt.show()
```

## Subplots Functionality

```python
a = np.linspace(0,10,11)
b = a**4
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
# axes is a numpy array when not 1x1. If n-d array, need to fix indexing of aces
axes[0][0].plot(x,y) # Only plotting one of the graphs.
axes[0][1].plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=3, ncols=1)
axes[0].plot(x,y)
fig.tight_layout() # Adjust spacing of graphs

# for ax in axes:
#     ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][0].set_ylabel('1, 0')
axes[1][1].plot(x,y)

fig.suptitle('Figure Level', fontsize=16)
fig.subplots_adjust(wspace=0.9, hspace=0.5) # More spacing adjustment. Spacing by axes width - if wspace=1 can fit another graph in between
```

## Styling

```python
# Legends
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = np.linspace(0,10,11)
ax.plot(x,x, label='X vs. X')
ax.plot(x,x**2, label='X vs. X^2')
ax.legend(loc=(1.1,0.5))
```

```python
# Colors and Styles
x = np.linspace(0,10,11)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# Line options w/ ls = '-' or '--' or '-.' or ':' or '.'. marker to see actual point you are plotting - adjust size w/ ms.
# markercolorface, markeredgewidth, markeredgecolor
lines = ax.plot(x,x, color='#650bdb', lw=1, ls='--', marker='o') # RGB Hex Code, lw for line width (how thick line is.)
lines[0].set_dashes([5,2,10,2]) # Short 5 dash, 2 white spaces, long dash 10, 2 white spaces
ax.plot(x, x+1, color='red', label= 'x vs. x+1')
ax.legend()
```
