```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("../UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/movie_scores.csv")
df.head()
```

```python
# KEEP DATA
df[(df['pre_movie_score'].isnull()) & (df['first_name'].notnull())]
```

```python
# DROP DATA
df.dropna(thresh=1) # Don't drop if they have at least 1 non-null value.
df.head()
```

```python
df.dropna(axis=1) # Drops everything because looks at if a column has a missing value. axis=0 for rows w/ at least 1 null
```

```python
df.dropna(subset=['last_name']) # Only considers this column having missing values.
```

```python
# FILLING DATA
df.fillna('new val') # Fills all missing values w/ inputted value
```

```python
df['pre_movie_score'].mean() # Gives mean of existing values
df['pre_movie_score'].fillna(df['pre_movie_score'].mean()) # Get mean per each column w/ existing values
```
