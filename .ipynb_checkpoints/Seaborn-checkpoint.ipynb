## Scatterplots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/dm_office_sales.csv")
df.head()
```

```python
# String need to match what's in df
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')#, palette='Dark2')
```

## Distribution Plots

```python
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
# sns.set(style='darkgrid')
sns.displot(data=df, x='salary', kde=True, bins=20, color='red', edgecolor='blue', linewidth=2, ls='--')
```

```python
# HISTPLOT
sns.histplot(data=df, x='salary')
```

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
sample_ages.head()
```

```python
sns.rugplot(data=sample_ages, x='age')
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
```

```python
# Cutoff KDE, bw_adjust for bandwidth
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.5, shade=True)
```

## Categorical Plots

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
# Same as value_counts() but in graph form.
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
# plt.ylim(90, 260) Decieves bc you think computer software looks like it has no employees.

plt.show()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')
plt.legend(bbox_to_anchor=(1.05,1))

plt.show()
```

## Distributions within Categories

```python
df = pd.read_csv("../UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/StudentsPerformance.csv")
df.head()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df,y='math score', x='parental level of education',
            hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.violinplot(data=df,x='reading score', y='parental level of education',
                palette='Set2')
# plt.legend(bbox_to_anchor=(1.05,1))
# Hard to understand this plot.
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.swarmplot(data=df,x='reading score', y='gender', size=2,
              hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.2,1))
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxenplot(data=df,x='reading score', hue='gender',
              y='test preparation course')
plt.legend(bbox_to_anchor=(1.2,1))
```

## Comparison Plots 

```python
sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(data=df, hue='gender')
```

## Grid Plots

```python
sns.catplot(data=df, x='gender', y='math score', kind='box',row='race/ethnicity', col='lunch')
```

```python
g = sns.PairGrid(data=df, hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
g = g.add_legend()
```

## Matrix Plots

```python
df = pd.read_csv("../UNZIP_FOR_NOTEBOOKS_FINAL/05-Seaborn/country_table.csv")
df.head()
```

```python
df = df.set_index('Countries')
df
```

```python
plt.figure(dpi=200)
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True,
            cmap='viridis', center=40)
```

```python
plt.figure(dpi=200)
sns.clustermap(data=df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True,
            cmap='viridis', col_cluster=False)
```
