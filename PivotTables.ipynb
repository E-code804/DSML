```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/Sales_Funnel_CRM.csv")
df.head()
```

```python
help(pd.pivot)
```

```python
# licenses of each product type purchased by google.
licenses = df[['Company', 'Product', 'Licenses']]
licenses
```

```python
pd.pivot(data=licenses, index='Company', columns='Product', values='Licenses')
```

```python
df.groupby('Company').describe()
```

```python
pd.pivot_table(df, index='Company', aggfunc='sum', values=['Licenses', 'Sale Price'])
```

```python
df.groupby('Company').sum()
```

```python
pd.pivot_table(df, index=['Account Manager', 'Contact', 'Product'], values=['Sale Price'],aggfunc=[np.sum], fill_value=0, margins=True)
```

```python
df.groupby(['Account Manager', 'Contact'], as_index=False)['Sale Price'].sum()
df.groupby(['Account Manager', 'Contact'], as_index=False).agg({'Sale Price': 'sum'})
```
