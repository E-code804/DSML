```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./UNZIP_FOR_NOTEBOOKS_FINAL/03-Pandas/tips.csv")
df.head()
```

```python
# Get last 4 digits of CC Number
def last_four(num):
    return str(num)[-4:]
df['last_four'] = df['CC Number'].apply(last_four)
```

```python
df['total_bill'].mean()
```

```python
def yelp(price):
    if price < 10:
        return '$'
    elif price < 30:
        return '$$'
    return '$$$'
df['yelp'] = df['total_bill'].apply(yelp)
```

```python
def quality(total_bill, tip):
    ratio = tip / total_bill
    if ratio > 0.25:
        return 'Generous'
    return 'Other'
```

```python
quality(16.99, 1.01)
```

```python
df['Quality'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis=1)
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
```
